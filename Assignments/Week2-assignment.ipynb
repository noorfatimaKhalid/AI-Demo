```python
#ASSIGNMENT 1: The Data Analyzer
def get_num(): #getting numbers
   l_num=list(map(int,input("enter a list of numbers").split()))
   return l_num
def get_mean(l_num): #finding mean
    sum=0
    for i in l_num:
        sum=sum+i
    mean=sum/len(l_num)
    return mean
def get_variance(l_num,mean): #finding variance
    variance = sum((x - mean)**2 for x in l_num)/(len(l_num)-1)
    return variance
def get_std(variance): #for standard derivation 
    return variance*0.5
list_num = get_num()  #storing nethod output into variables
mean = get_mean(list_num)
variance = get_variance(list_num, mean)
std = get_std(variance)
print('Data',list(map(float,list_num))) #printing data 
print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std)

```

```python
#ASSIGNMENT 2: The Student Class System
# Base class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)


# Derived class Student
class Student(Person):
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)


# Derived class GraduateStudent
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic

    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)
        print("Thesis Topic:", self.thesis_topic)


# Creating objects and calling methods
print("Person Info:")
p = Person("Ali", 30)
p.display_info()

print("\nStudent Info:")
s = Student("Ahmed", 20, "S123", "Computer Science")
s.display_info()

print("\nGraduate Student Info:")
g = GraduateStudent("Sara", 25, "G456", "Software Engineering", "AI in Healthcare")
g.display_info()
    
```

```python
#ASSIGNMENT 3: Shape Calculator with Magic Methods
class Rectangle :
    def __init__(self,length,width):
       self.length=length
       self.width=width
    def __str__(self): #returning rectangle
        return  f"Rectangle(width={self.width}, height={self.length})"
    def Area(self): #calculating area
        self.area=self.length * self.width
        return self.area
    def __add__(self,other): #overloading + to return new object
       new_length=self.length+other.length
       new_width=self.width+other.width
       return Rectangle(new_length, new_width)
    def __eq__(self,other): #comparing areas
         return self.Area()==other.Area()
rect1=Rectangle(4,5)
rect2=Rectangle(4,5)
print(rect1.Area())
print(rect2.Area())
rect3=rect1+rect2
print(rect3)
print("is equal" ,rect1==rect2)
```
