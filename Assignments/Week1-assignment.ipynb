```python
#Assignment 1: The Personal Assistant (Day 1-3 Basics)
print("What is your name? ")
name=input() #Ask the user for their name and store it in a variable. 
print("What year were you born?  ")
year=int(input()) #Ask the user for their birth year (as an integer) and store it.
print("Enter your current age :")
c_age=int(input()) #Ask the user for their current age (as an integer) and store it.
def cal_age(c_age,year): #Calculate and print the year they will turn 100 years old
   p_age=100-c_age+year
   return p_age
print("Enter your lucky number:")
l_num=float(input())  #Ask the user for a lucky number (as a float)
def show(): #Printing Summray
    print("Hello, ",name)
    print("You were born in ",year)
    print("You will turn 100 in the year ",cal_age(c_age,year))
    print("Your lucky number is ",l_num)
show()





```

```python
#Assignment 2: The Shopping List Manager (Day 4 Data Structures)
shopping_list=["milk",'eggs','apple','chicken']  #Create a list named shopping_lis



item_price={"milk":250.99,'eggs':180.89,'apple':300.50,'chicken':700.0}

def show():

   print('Initial shopping list is :',shopping_list) #print the entire list
   print('The third item is ',shopping_list[2])  #print third item
   shopping_list.append('cereal')  #adding two new items
   shopping_list.append('yogurt')
   print("Adding 'cheese' and 'coffee'....")
   print('Removing milk:',shopping_list.remove('milk')) #removing one
   print("Updated Shopping list :",shopping_list)
   print("item prices",item_price) #showing item price
   print('The price of Apple is',item_price['apple']) # showing price of apple
show()
```

```python
#Assignment 3: The Number Cruncher (Day 5 Loops and Conditionals)
for i in range(21):
    if i%3==0: #If divisible by 3
        print('Fizz')
    elif i%5==0 :  #If divisible by 5
        print("Buzz")
    elif (i%3==0) and (i%5==0): #f divisible by both
        print('fizzBuzz')
    else:
        print(i)
print("enter any number")  #ask the user for a number.
x=int(input())
while(x>=0):
    print('countdown',x)
    x=x-1 
    if x==0:
        print('BlastOff!')
        break
```
