```python
import pandas as pd
import numpy as np 
df=pd.read_csv("D:\Python Notebooks\sales.csv")
#Standardize 'region' values (ensure consistent capitalization and remove extra spaces)

df['region'].str.strip().str.capitalize()
df

```

```python
#Calculate total sales by region and find which region has the highest sales
df.groupby('region')['sales'].sum().idxmax()
#Find the average discount given for each product category
df.groupby('product')['discount'].mean()
#Analyze customer ratings - calculate average rating by customer type (New vs Returning)
#df.groupby('customer_type')['ratings'].mean()
#Find top 5 products by quantity sold
df.sort_values(by='quantity',ascending=False).head(5)['product']
#Calculate monthly sales trend by extracting month from date column (check pandas date time modules)
df['date']=pd.to_datetime(df['date'])
df.groupby(df['date'].dt.month)['sales'].sum()
#Analyze discount impact on sales - compare average sales with and without discount
#df.groupby('Discount')['Sales'].mean()
#Find the most profitable product (highest total sales amount)
df.groupby('product')['sales'].sum().idxmax()

#Analyze customer behavior - compare purchase patterns between New and Returning customers
df.groupby("customer_type")['sales'].mean()

```
