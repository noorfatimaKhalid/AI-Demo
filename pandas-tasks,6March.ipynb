```python
import numpy as np
import pandas as pd
#Importing file
df=pd.read_csv("D:\Python Notebooks\data.csv")
df
#Display the first 5 rows of the dataset
print(df.head(5))
#Display the last 5 rows.
print(df.tail(5)) 
#Check the shape of the dataset (rows and columns).
print(df.shape)
#Display the column names. 
print(df.columns)
#Check the data types of all columns
df.info()
#Generate a statistical summary of the dataset
df.describe()
#ind the maximum Pulse value. 
print(df.max())


```

DATA CLEANING

```python
#data colomn into datetime format
df['Date']= pd.to_datetime(df['Date'],errors='coerce')
df.info()
# Identify missing values in the dataset. 
df.isnull()
#Replace missing Calories values with the mean Calories.
df['Calories']= df['Calories'].fillna(df['Calories'].mean())
#Find and remove duplicate rows (there is a duplicate date). 
df.duplicated()
df.drop_duplicates()
# Detect incorrect Duration values (e.g., 450 minutes) and fix them.
df['Duration']= df['Duration'].replace(450,60)
#Remove rows where Date is missing. 
df.dropna(subset=['Date'])
#. Replace the date format "20201226" with proper datetime format. 
pd.to_datetime(df['Date'], format='Y%m%d')


```

DATA SELECTION TASKS

```python
#Select rows where Pulse > 110. 
df.loc[df['Pulse'] >110]
#Filter rows where Calories burned are greater than 350. 
df.loc[df['Calories'] > 350]
# Select rows where Pulse is between 100 and 110.
df.loc[(df['Pulse']>100) & (df['Pulse']<110 )]
#Display only Duration and Calories columns. 
df[['Calories','Duration']]
```

Data Analysis Tasks 

```python
#Calculate the average Pulse for the dataset. 
df['Pulse'].mean()
#Find the day with the highest Calories burned.
df.groupby('Date')['Calories'].sum().idxmax() #phle date ky groups bnae ga hr grp ka seperate sepeprate calories count kre ga aur max wali ko print kara dy ga
#  Find the total Calories burned in December. 
df.loc[(df['Date'].dt.month)==12,'Calories'].sum()
#. Calculate the average Duration of workouts. 
df['Duration'].mean()
# Find the maximum Maxpulse value. 
df['Maxpulse'].max()
```

5⃣ Sorting & Grouping Tasks

```python
#Sort the dataset by Calories burned (descending). 
df.sort_values(by='Calories',ascending=False)
#Sort the dataset by Pulse (ascending). 
df.sort_values(by='Pulse',ascending=True)
# Group the data by Duration and calculate average Pulse. 
df.groupby('Duration')['Pulse'].mean().sum()
# Find the top 5 workouts with the highest Calories burned.
df.sort_values(by='Calories',ascending=False).head(5)['Duration']

```
